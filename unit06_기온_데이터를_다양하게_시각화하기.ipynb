```python
# 주사위를 5번 던져서 나온 데이터를 최종 결과로 한 번에 출력하기

import random

dice=[] # 주사위 리스트 변수 선언 : 순서대로 데이터를 담기 위해

for i in range(5):
  dice.append(random.randint(1,6))

dice

# 랜덤하게 굴린 주사위 5개의 값을 히스토그램으로 표현
plt.figure(dpi=200)
plt.hist(dice) # 구간을 지정해주지 않으면 범위를 제멋대로 지정할 가능성이 있다
plt.show()
```

최고 기온 데이터를 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

result=[] # 최고 기온 데이터를 저장할 리스트

for row in data:
  if row[-1] != '': # 최고 기온 데이터 값이 결측치가 아니라면
    result.append(float(row[-1]))

# 최고 기온 데이터 히스토그램
plt.figure(dpi=300)
plt.hist(result, bins=100,color='r')
plt.show()
```

```python
# 최고 기온 데이터 히스토그램
plt.figure(dpi=300)
plt.hist(result, bins=20,color='r')
plt.show()
```

1월과 8월의 데이터를 히스토그램으로 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

aug=[] # 8월의 최고 기온 데이터를 저장할 리스트 선언
jan=[] # 1월의 최고 기온 데이터를 저장할 리스트 선언

for row in data:
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터를 추출하여 month변수에 저장
  
  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))
    elif month == '01':
      jan.append(float(row[-1]))

# 최고 기온 데이터 히스토그램
plt.figure(dpi=300)
plt.hist(aug, bins=100,color='r',label='Aug')
plt.hist(jan, bins=100,color='b',label='Jan')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

aug=[] # 8월의 최고 기온 데이터를 저장할 리스트 선언
jan=[] # 1월의 최고 기온 데이터를 저장할 리스트 선언
apr=[] # 4월의 최고 기온 데이터를 저장할 리스트 선언

for row in data:
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터를 추출하여 month변수에 저장
  
  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))
    elif month == '01':
      jan.append(float(row[-1]))
    elif month == '04':
      apr.append(float(row[-1]))

# 최고 기온 데이터 히스토그램
plt.figure(dpi=300)
plt.hist(aug, bins=100,color='r',label='Aug')
plt.hist(jan, bins=100,color='b',label='Jan')
plt.hist(apr, bins=100,color='g',label='Apr')
plt.legend()
plt.show()
```

1월부터 12월까지 최고기온 데이터를 상자그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

month=[[],[],[],[],[],[],[],[],[],[],[],[]] # 월별 데이터 저장할 month변수 선언

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) 
    # 월과 같은 번호의 인덱스에 월별 데이터 저장 (ex. month의 0번째 방에는 1월 데이터)
  
plt.figure(dpi=200)
plt.boxplot(month)
plt.show() # 주황색은 중앙값

```

8월의 일별 최고 기온 데이터를 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

day=[]

for i in range(31):
  day.append([]) # day리스트 내에 31개의 리스트 생성
  # day = [[] for i in range(31)]

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))
    # 8월 최고 기온 값 저장 day[0] -> 1일

plt.figure(figsize=(10,5),dpi=200) # 그래프의 길이를 조정하여 데이터가 깨지지 않도록
plt.boxplot(day, showfliers=False) # outlier(이상치) 제거
plt.show() # 주황색은 중앙값
```

스타일 변형

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

day=[]

for i in range(31):
  day.append([]) # day리스트 내에 31개의 리스트 생성
  # day = [[] for i in range(31)] 이런 방법도 있음

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))
    # 8월 최고 기온 값 저장 day[0] -> 1일

plt.style.use('ggplot') # 그래프 스타일 지정(회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)

plt.figure(figsize=(10,5),dpi=200) # 그래프의 길이를 조정하여 데이터가 깨지지 않도록
plt.boxplot(day, showfliers=False) # outlier(이상치) 제거
plt.show() # 주황색은 중앙값
```
