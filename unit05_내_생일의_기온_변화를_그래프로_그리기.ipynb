데이터 준비하기

```python
# 최고 기온 데이터만 리스트에 출력하기
 
import csv

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

result=[] # 최고 기온 데이터를 저장할 리스트

for row in data:
  if row[-1] != '': # 최고 기온 데이터 값이 결측치가 아니라면
    result.append(float(row[-1]))

print(result) # 최고 기온 리스트 출력
```

최고기온 값 중 8월에 해당되는 값만 추출해서 그래프 시각화

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

result=[] # 최고 기온 데이터 중 8월의 데이터만 저장할 리스트

for row in data:
  if row[-1] != '': # 최고 기온 결측치 제외
    if row[0].split('-')[1] == '08': # 날짜 데이터를 잘라 8월 데이터만 추출
      result.append(float(row[-1])) # 결과 리스트에 담기

plt.figure(dpi=200) # 해상도 조절
plt.plot(result,'yellowgreen')
plt.show()

```

마이너스(-) 깨짐 처리하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

high=[]
low=[]

for row in data:
  if row[-1] != '' and row[-2] != '':
    if int(row[0].split('-')[0]) >= 1992:
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '24':
        high.append(float(row[-1]))
        low.append(float(row[-2]))

# 데이터 시각화

#차트 설정
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# plot 차트 그리기
plt.plot(high,'lightpink',label='최고기온')
plt.plot(low,'skyblue', label='최저기온') # 한글 폰트를 씌웠기 때문에 마이너스(-)값에 대한 오류 발생

# 마무리 설정(제목, 범례, 보여주기 명령어 등 디테일을 위한 조건(필수조건은 아님))
plt.title('내 생일의 기온변화 그래프')
plt.legend() # 범례 표시
plt.show()

```

매년 생일 기준 최고 기온, 평균 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)

high=[]
low=[]
mean=[]

for row in data:
  if row[-1] != '' and  row[-2] != '' and row[-3] != '': # 최고, 최저, 평균 결측치 제거
    if int(row[0].split('-')[0]) >= 1992:
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '24':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        mean.append(float(row[-3]))

# 데이터 시각화

#차트 설정
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# plot 차트 그리기
plt.plot(high,'lightpink',label='최고기온')
plt.plot(low,'skyblue', label='최저기온') # 한글 폰트를 씌웠기 때문에 마이너스(-)값에 대한 오류 발생
plt.plot(mean,'yellowgreen', label='평균기온')
# 마무리 설정(제목, 범례, 보여주기 명령어 등 디테일을 위한 조건(필수조건은 아님))
plt.title('내 생일의 기온변화 그래프')
plt.legend() # 범례 표시
plt.show()
```
