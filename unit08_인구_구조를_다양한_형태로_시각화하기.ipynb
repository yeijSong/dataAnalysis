**항아리 모양 그래프 그리기**

성별 데이터 저장하기(Type1) : dataset이 바뀔 가능성이 있을 때 더 효율적

```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

m=[]
f=[]

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101) : # 컬럼의 개수를 알고 있을 때
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)])) # -1 값부터 담음 -> 여성100세 이상부터 담음

f.reverse() # 뒤에서부터 담았으니 다시 0세부터 출력되도록 reverse함

print(m)
print(f)
```

성별 데이터 순서대로 저장하기(Type2) : 직관적임, data가 바뀔 가능성이 없을 때 좋음

```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

m=[]
f=[]

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]: # 남성데이터(3~103번 데이터 슬라이싱)
      m.append(int(i))
    for i in row[106:]: # 여성데이터(106~끝까지 데이터 슬라이싱)
      f.append(int(i))

print(m)
print(f)
```

데이터 시각화하기

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()

# 겹쳐져 있으니 비교불가
```

항아리 그래프를 위해 남성데이터 음수로 바꾸기

```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

m=[]
f=[]

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]: # 남성데이터(3~103번 데이터 슬라이싱)
      m.append(-int(i))
    for i in row[106:]: # 여성데이터(106~끝까지 데이터 슬라이싱)
      f.append(int(i))

print(m)
print(f)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False
plt.barh(range(101),m,color='navy',label='남성')
plt.barh(range(101),f,color='pink',label='여성')
plt.legend()
plt.title('신도림 지역의 남녀 성별 인구분포')
plt.show()
```

2) 2019년 2월 연령별 인구구조 -> 2020년 2월 연령별 인구구조 비교(같이 그리기)


```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

m=[]
f=[]

for row in data:
  if '서구 내동' in row[0]:
    for i in row[3:104]: # 남성데이터(3~103번 데이터 슬라이싱)
      m.append(-int(i))
    for i in row[106:]: # 여성데이터(106~끝까지 데이터 슬라이싱)
      f.append(int(i))

print(m)
print(f)
```
