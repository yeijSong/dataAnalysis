꺾은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

m=[]
f=[]

name=input('지역명을 입력하세요 : ')
for row in data:
  if name in row[0]:
    for i in row[3:104]: # 남성데이터(3~103번 데이터 슬라이싱)
      m.append(int(i.replace(',','')))
    for i in row[106:]: # 여성데이터(106~끝까지 데이터 슬라이싱)
      f.append(int(i.replace(',','')))
    break


plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False

# 꺾은선 그래프로 남녀 인구수 데이터 표현하기

plt.plot(m, 'skyblue', label='남성')
plt.plot(f, 'pink', label='여성')
plt.legend()
plt.title(name+'의 성별 인구 구조')
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기

```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

result=[] # 남성 인구수 - 여성 인구수

name=input('지역명을 입력하세요 : ')
for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False

plt.bar(range(101),result, color='green')

plt.title(name+'의 남녀 성별 인구분포')
plt.show()

```

**위치, 크기가 서로 다른 100개의 점 만들기**

### 원의 크기를 조절하여 자료를 효과적으로 표현하기

- 인구수의 제곱근을 추가하여 원의 크기 조정
- 인구수는 데이터의 크기가 매우 크기때문에 제곱근을 사이즈로 하는 것임

```python
# 인구수에 제곱근을 적용하여 사이즈를 준 산점도

import csv
import matplotlib.pyplot as plt
import math

f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

m=[]
f=[]
size=[]

name=input('지역명을 입력하세요 : ')
for row in data:
  if name in row[0]:
    for i in row[3:104]: # 남성데이터(3~103번 데이터 슬라이싱)
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',',''))))
    for i in row[106:]: # 여성데이터(106~끝까지 데이터 슬라이싱)
      f.append(int(i.replace(',','')))
    break

plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False

# 산점도 그래프로 남녀 인구수 데이터 표현하기

plt.scatter(m,f,s=size,c=range(101),cmap='jet',alpha=0.7) # 남성의 인구수를 기준으로 사이즈 추가
plt.plot(range(max(m)),range(max(m)),'g')
plt.colorbar() # 연령과 관련된 정보를 구체적으로 알기위해 color와 colorbar추가
plt.show()

# 내가 더 강조해서 보고싶은 자료를 x축에 두면 더 보기 쉽다
# 인원수가 많은 구간은 점을 크게하여 직관적으로 자료를 파악할 수 있도록 했음
```
