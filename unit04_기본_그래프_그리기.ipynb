plt.plot(dataset) 그래프 그리기

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.plot([10, 20, 30, 40]) # y값이 디폴트(머신러닝에서 중요한 것은 y값)
plt.show()

#  x값은 자동으로 0부터 1씩 증가하는 형태(데이터의 크기에 따라 x값의 범위가 변경되기도 함)
```

```python
import matplotlib.pyplot as plt

plt.title('plotting')
plt.plot([10,20,30,40],label='asc') # asc=ascending을 의미
plt.plot([40,30,20,10],label='desc') # desc=descending을 의미

plt.legend() # 그래프의 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('plotting')
plt.plot([10,20,30,40],label='asc') # asc=ascending을 의미
plt.plot([40,30,20,10],label='desc') # desc=descending을 의미

plt.legend(loc=10) # 그래프의 범례 속성 출력 위치 이동
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('color')

plt.plot([10,20,30,40],color='skyblue',label='skyblue') # 그래프의 색 지정
plt.plot([40,30,20,10],'pink',label='pink') # 색 이름만 써도 된다 but 명확한 코딩을 위해 그냥 color써줄 것

plt.legend() # 그래프의 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('color')

plt.plot([10,20,30,40],color='r',label='red') # 기본 색상은 약자로 가능 (r-red, b-blue, g-green, y-yellow, k-black 등)
plt.plot([40,30,20,10],color='k',label='black') # 색 이름만 써도 된다 but 명확한 코딩을 위해 그냥 color써줄 것

plt.legend() # 그래프의 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('linestyle')

plt.plot([10,20,30,40],color='skyblue',linestyle='--',label='dashed') # 그래프의 선 스타일 지정
plt.plot([40,30,20,10],'pink',ls=':',label='dotted') # ls=linestyle, ':'는 좀 더 자잘한 점선

plt.legend() # 그래프의 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('marker')

plt.plot([10,20,30,40],'r.',label='circle') # 마커를 빨간색 동그라미 모양
plt.plot([40,30,20,10],'k^',label='triangle up') # 마커를 검정색 삼각형 모양으로

plt.legend() # 그래프의 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('marker')

plt.plot([10,20,30,40],'r.-',label='circle') # '색상,마커모양,선모양' 코드로 작성 가능
plt.plot([40,30,20,10],'k^--',label='triangle up')

plt.legend() # 그래프의 범례 속성 추가
plt.show()
```

데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f)
header=next(data)

result=[] # 결측치를 제거한 서울시의 최고 기온 리스트

for row in data:
  if row[-1] != '': # 최고 기온이 ''(빈문자열)이 아니면,
    result.append(float(row[-1])) # 값을 실수로 변환하여 리스트에 추가

result

plt.plot(result,'r')
plt.show
```

```python
plt.figure(figsize=(10,2),dpi=200)
plt.plot(result,'r')
plt.show
```
