```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv',encoding='cp949')
data=csv.reader(f)

result=[] # 순서대로 데이터를 담기위하여 리스트에 데이터 저장

for row in data:
  if '서구' in row[0] and '내동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',',''))) # 인구 수가 1000명 이상이면 데이터에 ','가 있으므로 없애주고 int

plt.figure(dpi=200)
plt.style.use('ggplot') # 격자무늬 스타일로 지정
# 격자 무늬 스타일로 지정 classic, dark_background,ggplot
plt.plot(result)
plt.title('대전광역시 서구 내동 지역의 인구 구조')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv',encoding='cp949')
data=csv.reader(f)

result=[] # 순서대로 데이터를 담기위하여 리스트에 데이터 저장

for row in data:
  if row[0].split()[0]== '대전광역시' and row[0].split()[1]== '서구' and '내동' in row[0].split()[2]:
    for i in row[3:]:
      result.append(int(i))

plt.figure(dpi=200)
plt.style.use('ggplot') # 격자무늬 스타일로 지정
# 격자 무늬 스타일로 지정 classic, dark_background,ggplot
plt.plot(result)
plt.show()
```
