**신도림 데이터 확인하기**

in 연산자를 활용하여 우리동네 이름이 포함된 지역명 찾기

```python
'신도림' in '서울특별시 구로구 신도림동(1153051000)'
```

in 연산자를 활용하여 우리동네의 데이터 출력하기 -> 우리동네 데이터인지 확인

```python
import csv

f=open('age.csv',encoding='cp949')
data=csv.reader(f)

for row in data:
  if '신도림' in row[0]:
    print(row)
```

연령별 인구수(0~100세 이상) 데이터 확인

```python
import csv

f=open('age.csv',encoding='cp949')
data=csv.reader(f)

for row in data:
  if '신도림' in row[0]:
    print(row[3:]) # 0~100세 이상 인구수 출력
    # row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~ : 연령별 인구 수
    print()
    print(len(row[3:])) # 0~100세 이상 데이터 수 출력 -> 101개
```

우리 동네 연령별 인구수 result리스트에 저장하기 (for 시각화)

```python
import csv

f=open('age.csv',encoding='cp949')
data=csv.reader(f)

result=[] # 순서대로 데이터를 담기위하여 리스트에 데이터 저장

for row in data:
  if '신도림' in row[0]:
    for i in row[3:]:
      result.append(int(i))

result
```

우리 동네 연령별 인구수 데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.style.use('ggplot') # 격자무늬 스타일로 지정
# 격자 무늬 스타일로 지정 classic, dark_background,ggplot
plt.plot(result)
plt.show()
```
