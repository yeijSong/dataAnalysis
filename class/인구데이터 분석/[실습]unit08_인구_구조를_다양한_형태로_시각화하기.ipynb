```python
import csv
import matplotlib.pyplot as plt

f=open('2011pop.csv',encoding='cp949')
data=csv.reader(f)

f2=open('17age09.csv',encoding='cp949')
data2=csv.reader(f2)

result=[]
result2=[]

for row in data:
  if '노량진제1동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))

for row in data2:
  if '노량진제1동' in row[0]:
    for i in row[3:]:
      result2.append(int(i.replace(',','')))


plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')
plt.title('2017년 노량진제1동 지역의 인구 구조')
plt.plot(result,'r',label='3월')
plt.plot(result2,'b',label='9월')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('17age03.csv',encoding='cp949')
data=csv.reader(f)

f2=open('17age09.csv',encoding='cp949')
data2=csv.reader(f2)

result=[]
result2=[]

for row in data:
  if '노량진제1동' in row[0]:
    for i in row[23:44]:
      result.append(int(i.replace(',','')))

for row in data2:
  if '노량진제1동' in row[0]:
    for i in row[23:44]:
      result2.append(int(i.replace(',','')))


plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')
plt.title('2017년 노량진제1동 지역의 20~40세 인구 구조')
plt.plot([i for i in range(20,41)],result,'r',label='3월')
plt.plot([i for i in range(20,41)],result2,'b',label='9월')
plt.legend()
plt.show()
```

막대그래프 그리기 ->비교 (지역별, 성별)

```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv',encoding='cp949')
data=csv.reader(f)

result=[] # 순서대로 데이터를 담기위하여 리스트에 데이터 저장

for row in data:
  if '월평3동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))

# 인구 구조 시각화

plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.title('월평3동 지역의 인구 구조')
plt.bar(range(101),result) #x축과 y축을 정의하지 않으면 오류 발생
plt.show()
```

수평 막대그래프(barh())

```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv',encoding='cp949')
data=csv.reader(f)

result=[] # 순서대로 데이터를 담기위하여 리스트에 데이터 저장

for row in data:
  if '월평3동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))

# 인구 구조 시각화

plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.title('월평3동 지역의 인구 구조')
plt.barh(range(101),result) #x축과 y축을 정의하지 않으면 오류 발생
plt.show()
```
