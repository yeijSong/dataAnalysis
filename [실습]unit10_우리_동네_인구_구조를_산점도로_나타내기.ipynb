꺾은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

m=[]
f=[]

name=input('지역명을 입력하세요 : ')
for row in data:
  if name in row[0]:
    for i in row[3:104]: # 남성데이터(3~103번 데이터 슬라이싱)
      m.append(int(i.replace(',','')))
    for i in row[106:]: # 여성데이터(106~끝까지 데이터 슬라이싱)
      f.append(int(i.replace(',','')))
    break


plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False

# 꺾은선 그래프로 남녀 인구수 데이터 표현하기

plt.plot(m, 'skyblue', label='남성')
plt.plot(f, 'pink', label='여성')
plt.legend()
plt.title(name+'의 성별 인구 구조')
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기

```python
import csv
import matplotlib.pyplot as plt


f=open('gender.csv',encoding='cp949')
data=csv.reader(f)

result=[] # 남성 인구수 - 여성 인구수

name=input('지역명을 입력하세요 : ')
for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=200)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False

plt.bar(range(101),result, color='green')

plt.title(name+' 지역의 남녀 성별 인구분포')
plt.show()

```
